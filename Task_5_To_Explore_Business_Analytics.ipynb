**Visualizing different numeric variables of the dataset** 

```
print(df['Profit'].describe())
plt.figure(figsize=(9, 8))
sns.distplot(df['Profit'], color='g', bins=100, hist_kws={'alpha': 0.4});
```

Profit data is normally distributed

```
print(df['Discount'].describe())
plt.figure(figsize=(9, 8))
sns.distplot(df['Discount'], color='g', bins=100, hist_kws={'alpha': 0.4});
```

Discount data is randomly distributed

```
print(df['Sales'].describe())
plt.figure(figsize=(9, 8))
sns.distplot(df['Sales'], color='g', bins=100, hist_kws={'alpha': 0.4});
```

Sales data is a bit skewed to the right.

```
print(df['Quantity'].describe())
plt.figure(figsize=(9, 8))
sns.distplot(df['Quantity'], color='g', bins=100, hist_kws={'alpha': 0.4});
```

Quantity data is normally distributed

```
#To find the relation between numerical variables in our dataset
df.corr()
```

```
#the product categories available to shop for customers are:
df['Category'].unique()
```

```
#number of products in each category are:
df['Category'].value_counts()
```

```
#number of Sub-categories products divided are:
df['Sub-Category'].nunique()
```

```
#Lets see how sub-categories are distributed with respect to category
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category',data=df,color='purple')
plt.show()
```

From this graph, one can easily makeout which Category & Sub-Category to choose when they are looking to purchase a product

```
plt.figure(figsize=(12,10))
df['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

Pie Chart shows that the store has wide variety of Office Supplies especially in Binders and Paper department.

```
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
plt.show()
```

Selling price for Chairs and Phones is extremely high compared to other products.
Highest profit is earned in Copiers.

```
#Count of Sub-Category region wise
plt.figure(figsize=(15,8))
sns.countplot(x="Sub-Category", hue="Region", data=df)
plt.show()
```

```
df['Cost']=df['Sales']-df['Profit']
df['Cost'].head()
```

```
df['Profit %']=(df['Profit']/df['Cost'])*100
```

```
df.sort_values(['Profit %','Sub-Category'],ascending=False).groupby('Profit %').head(5)
```

From the above table we can infer that by selling phones and papers retailers got 100percent profit.

Visualizing different classes

```
df['Ship Mode'].value_counts()
```

```
sns.countplot(x='Ship Mode', data=df)
plt.title('No of sales made per shipping mode')
```

Sales per shipping mode is highest in standard class

```
df.groupby(['Ship Mode'])['Sales','Profit'].sum().plot.bar(figsize=(10,5),
           title='Sales and Profits per Ship Mode',color=['maroon','blue'])
```

The maximum sales amount and profit is made when items are shipped through the Standard Class shipping mode.

Top 10 cities with maximum profits

```
pd.pivot_table(df,index=['City'],values=['Profit'],aggfunc=np.sum).sort_values(by='Profit',ascending=False).head(10)
```

10 cities under loss

```
pd.pivot_table(df,index=['City'],values=['Profit'],aggfunc=np.sum).sort_values(by='Profit',ascending=False).tail(10)
```

**Conclusion:**

Some products like fastners arent being sold well in any region so retailers can stop selling such products.

The retailers have to be region specific. That is they have to sell particular products that have more demand in a particular region. 

Regions which have high demand of a particular product can have more discounts whereas regions with lower demands of that product can have less discounts.
